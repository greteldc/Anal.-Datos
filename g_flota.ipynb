## 3. Limpieza del dataframe

### 3.1. Selección de los datos relevantes

a) las columnas relevantes para nuestro propósito:

```python
flota = flota.iloc[:,[0,7,8]]
#acc_2020_col = acc_2020.iloc[:,[0,2,6,12,15,18,21,24,27,29,30,31,32,33,34]]
flota
```

b) Selección de las filas relevantes para nuestro propósito:

```python
flota = flota.loc[[11,15,19,20,21,22,23], :]
flota
```

### 3.2. Limpieza layout

```python
#prever índice "limpio"
flota = flota.reset_index(drop = "index")
flota
```

```python
#rename columnas seleccionadas
flota = flota.rename(columns={"STATISTIEK VAN DE MOTORVOERTUIGEN OP 1 AUGUSTUS 2021": "Year", "Unnamed: 7": "vehículos", "Unnamed: 8": "motores y motocicletas" })
flota
```

## 4. Operaciones para obtener el nº total de unidades en la flota

### 4.1. Crear nueva columna con el nº total de unidades en la flota (vehículos + motores + motocicletas)

```python
flota["Total_flota"] = flota["vehículos"] + flota["motores y motocicletas"]
flota
```

### 4.2. Limpieza del layout

```python
#limpieza después de la adición de columna:
flota = flota.drop(["vehículos","motores y motocicletas"], axis=1)
flota
```

## 5. Preparación de los ficheros relevantes para su exportación

```python
flota.to_csv("Data_ACC/flota.csv", index=False)
```
