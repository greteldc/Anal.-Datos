### I.3.2. Tratamos los valores no disponibles (emergidos en el I.2)

a) Identificamos las columnas concretas dónde se sitúan los valores no disponibles de una manera más visible que por el método .info() en el I.2.

```python
#guardamos los últimos cambios en la variables pobl_2019
pobl_2019 = pobl_2019_col
pobl_2019_col.isnull().any()
```

b) Apreciamos la cantidad de valores no disponibles (nº de registros en los cuales se manifiestan)

```python
pobl_2019.isnull().sum()
```

c) Con vistas a la sustitución de estos valores: verificamos primero si el número de 21492 corresponde al número de filas que conciernan solo al arrondissement Brussel-Hoofdstad

```python
#primer sondeo mediante sample:
nan_rows = pobl_2019[pobl_2019.isnull().any(1)]
nan_rows.sample(5)
```

```python
#cuento exacto:
sum(pobl_2019['ADM_DSTR'] == 'Arrondissement Brussel-Hoofdstad')
```

Conclusión: el número de filas sin valor disponible corresponde exactamente al número de filas que contienen "Arrondissement Brussel-Hoofdstad", ni más, ni menos. Por lo tanto, sabemos con certeza que todas las sustituciones se situarán en este "arrondissement". Para no perder datos en posteriores visualizaciones o tratamientos de datos en base a 'provincia', vamos a 'copiar' los valores de arrondissement a la columna de provincia. Así, siempre que llamemos a este dataframe (o de manera más concreta, a sus provincias), estaremos seguros de que no hay ningún valor NaN, y entrarán en consideración siempre los datos de la capital, aunque la 'etiqueta' no pueda considerarse propiamente como una provincia.

d) Sustituimos todos los valores "NaN" por "Arr. Brussel-Hoofdstad[no Prov]"

```python
pobl_2019['PROV'] = pobl_2019['PROV'].fillna("Arr. Brussel-Hfdstad[no Prov]")
```

```python
#verificamos si ahora nuestro dataframe se ha quedado sin valores NaN
pobl_2019.isnull().sum()
```

Ya está limpio pobl_2019: no quedan más valores Nan.

### I.3.3. Agrupamos las filas por 'provincia' y 'arrondissement' para obtener así el nº total de habitantes por 'arrondissement'. Optamos por mantener las provincias para que todos puedan situar con más facilidad los 'arrondissmenten'.

### a) agrupación de número de habitantes a nivel de 'arrondissement' 

```python
pobl_2019_habxarr = pobl_2019.groupby(["PROV","ADM_DSTR"])[["PROV","ADM_DSTR", "MS_POPULATION"]].sum()
pobl_2019_habxarr
```

```python
#reset index para merge en base a columnas idénticas
pobl_2019_habxarr.reset_index(inplace=True)
```

```python
#verificación: nuevo índice? sí
pobl_2019_habxarr
```

### b) agrupación de nº de habitantes por provincia (Bxl: arrondissment)

```python
pobl_2019_habxprov = pobl_2019_habxarr.groupby("PROV")[["PROV","MS_POPULATION"]].sum()
pobl_2019_habxprov
```

### c) agrupación de nº de habitantes a nivel de país

```python
hab_2019_BE = pobl_2019["MS_POPULATION"].sum()
hab_2019_BE
```

### d) nº de habitantes por "arrondissement": cifras absolutas y proporcionales respecto a la totalidad de la población belga

```python
pobl_2019_habxarr["PROPORTION_BE"] = pobl_2019_habxarr["MS_POPULATION"]/pobl_2019_habxarr["MS_POPULATION"].sum()
#por haber querido cambiar título nueva columna de PROPORCIÓN a PROPORCIÓN_BE, quitar la penúltima columna PROPORCION
#luego lo mismo con PROPORCIÓN_BE > PROPORTION_BE
#pobl_2019_habxarr.drop("PROPORCIÓN", axis=1, inplace=True)
#pobl_2019_habxarr.drop("PROPORCIÓN_BE", axis = 1, inplace=True)
pobl_2019_habxarr
```

### e) nº de habitantes por "provincia": cifras absolutas y proporcionales respecto a la totalidad de la población belga

```python
pobl_2019_habxprov["PROPORCIÓN_BE"] = pobl_2019_habxprov["MS_POPULATION"]/pobl_2019_habxprov["MS_POPULATION"].sum()
pobl_2019_habxprov
```
